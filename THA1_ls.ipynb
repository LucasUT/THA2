```python
#US Mass Shootings, 1982-2019: Data From Mother Jones’ Investigation

mjdata = pd.read_excel('https://docs.google.com/spreadsheets/d/1b9o6uDO18sLxBqPwl_Gh9bnhW-ev_dABH83M5Vb5L8o/export?format=xlsx&id=1b9o6uDO18sLxBqPwl_Gh9bnhW-ev_dABH83M5Vb5L8o')
print(mjdata.columns)
print(mjdata.shape)
mjdata.head(5)
```

```python
#Center for Homeland Defense and Security - School Shootings 

k12data = pd.read_csv('k12.csv',  encoding='latin-1')
print(k12data.columns)
print(k12data.shape)
k12data.head(1)
```

```python
#The data was downloaded from gunviolencearchive.org, https://www.kaggle.com/jameslko/gun-violence-data/home

gvdata = pd.read_csv('gun-violence-data.csv')
print(gvdata.columns)
print(gvdata.shape)
gvdata.head(1)
```

```python
print(list(mjdata))
print()
print(list(k12data))
```

```python
k12data.columns = map(str.lower, k12data.columns)
print(list(k12data))
print()
print(list(mjdata))
```

```python
newcol=[]
for ren in mjdata.columns:
    new = ren.replace('location','city')
    new = new.replace('age_of_shooter','shooter age')
    new = new.replace('gender','shooter gender')
    new = new.replace('race','shooter race')
    new=new.replace('weapon_type','firearm type')
    new=new.replace('weapon_details','caliber of firearm')
    new=new.replace('injured','wounded')
    new=new.replace('total_victims','total injured/killed victims')
    new=new.replace('location.1','school')
    new=new.replace('case','location')
    new=new.replace('fatalities','killed (includes shooter)')
    newcol.append(new)
mjdata.columns=newcol
mjdata.columns
```

```python
k12data.drop(['reliability score (1-5)', 'gender of victims (m/f/both)', "victim's affiliation w/ school", "victim's age(s) ", 'victims race', 'victim ethnicity', 'targeted specific victim(s)', 'bullied (y/n/ n/a)', 'domestic violence (y/n)', 'suicide (shot self immediately following initial shootings) y/n/ n/a', 'suicide (e.g., shot self at end of incident - time period between first shots and suicide, different location, when confronted by police) y/n/ n/a', 'suicide (or attempted suicide) by shooter (y/n)', "shooter's actions immediately after shots fired", 'pre-planned school attack (columbine style; not domestic)', 'category', 'school type', 'narrative (detailed summary/ background)', 'sources', 'time of first shot (24h)', 'during school day (y/n)', 'time period', 'during a sporting event (y/n)', 'during a school sponsored event (school dance, concert, play, activity)', 'number of shots fired', "shooter's affiliation with school", 'shooter had an accomplice who did not fire gun (y/n)', 'hostages taken (y/n)'],axis=1,inplace=True)
k12data.columns
```

```python
mjdata.drop(['latitude','longitude','type','mental_health_sources','sources_additional_age','weapons_obtained_legally','where_obtained'],axis=1,inplace=True)
mjdata.columns
```

```python
k12data.drop(['random victims','suicide (shooter was only victim) y/n/ n/a','duration (minutes)','day of week (formula)'], axis=1,inplace=True)
k12data.columns
```

```python
mj_k12 = pd.concat([mjdata,k12data], axis=0, ignore_index=True)
```
